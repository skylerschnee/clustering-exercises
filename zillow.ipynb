```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import env
```

```python
sql_query = '''
    select * 
    from properties_2017
    left join airconditioningtype using(airconditioningtypeid)
    left join architecturalstyletype using(architecturalstyletypeid)
    left join buildingclasstype using(buildingclasstypeid)
    left join heatingorsystemtype using (heatingorsystemtypeid)
    left join propertylandusetype using(propertylandusetypeid)
    left join storytype using(storytypeid)
    left join typeconstructiontype using(typeconstructiontypeid)
    left join unique_properties using(parcelid);
    '''
```

```python
def get_zillow_data():
    '''
    This function uses a SQL query 
    and get_connection function from the envy.py file
    to create df of zillow data
    '''    
    df = pd.read_sql(sql_query, env.get_connection('zillow'))
    
    return df
```

```python
def missing_values(df):
    num_rows = len(df)
    missing = df.isnull().sum()
    percent_missing = (missing / num_rows) * 100
    result = pd.concat([missing, percent_missing], axis=1)
    result.columns = ['num_rows_missing', 'pct_rows_missing']
    result.index.name = 'feature'
    return result.sort_values(by='pct_rows_missing', ascending=False)
```

```python
def handle_missing_values(df, prop_required_column, prop_required_row):
    '''
    The input:
    - A dataframe
    - A number between 0 and 1 that represents the proportion, for each column, 
    of rows with non-missing values required to keep the column. 
    - A number between 0 and 1 that represents the proportion, for each row, 
    of columns/variables with non-missing values required to keep the row. 
    
    The output:
    - The dataframe with the columns and rows dropped as indicated.
    '''
    # drop columns that do not meet prop_required_column
    df = df.dropna(thresh=int(prop_required_column*len(df)), axis=1)

    # drop rows that do not meet prop_required_row
    df = df.dropna(thresh=int(prop_required_row*len(df.columns)), axis=0)

    return df
```
